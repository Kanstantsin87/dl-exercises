```python
import linear_model as lm
from pprint import pprint
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
# чтение файла с данными
with open('home.txt') as f:
    # data - список из строк файла
    data = f.readlines()

def proc(string):
    # функция обработки строки с названием колонки
    return string.split(',')[1].strip().replace(';', '')

# создание списка с именами столбцов
columns = [proc(data[i]) for i in range(42,54)]
columns[0] += ' in hundreds of dollars'

# создание структуры pandas DataFrame
df = pd.DataFrame([data[i].split()[1:] for i in range(54,82)], columns=columns)

# запись в csv файл
df.to_csv('train.csv', index=False)
```

```python
data=np.genfromtxt("train.csv", skip_header=1, delimiter=",", dtype=np.float32)
data[:5]
```

```python

X = data[:, 0:11] 
y = data[:, 11]

print(X[0])
print(y[0])
```

```python
clf = lm.LinearRegression()
```

```python
clf.fit(X, y)
```

```python

print(clf.cost(X, y))
```

```python

plt.figure(figsize=(10, 7))
plt.scatter(X[:, 3], y)
plt.scatter(X[:, 3], clf.predict(X))
plt.xlabel('Жилая площадь (тыс. кв.футов)', fontsize=12)
plt.ylabel('Стоимость', fontsize=12)
plt.show()
```
